```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
rng = np.random.default_rng()
a_values=rng.integers(low=0,high=100,size=10)
b_values=rng.integers(low=100,high=500,size=10)
data= pd.DataFrame({'A':a_values,'B':b_values})
data
```

```python
#Crear una columna 
data['A_cat']=pd.cut(data.A, [0,30,40,50,60,100],labels=['muy poco','poco','normal', 'mucho', 'demasiado'])
data
```

```python
#mapear datos

A_cat_mapper ={'poco':'insuficiente',
               'muy poco':'insuficiente',
               'normal':'suficiente',
                'mucho': 'suficiente',
                'demasiado':'suficiente'}
```

```python
#Diccionario
data ['A_cat_suf_insuf']=data.A_cat.map(A_cat_mapper)
data
```

```python
data['A_cat_suf_insuf']= data.A_cat.map(A_cat_mapper)
data
```

```python
A_categories_1_dummies=pd.get_dummies(data.A_cat,drop_first=True,prefix ='A_cat1',dtype=int)
A_categories_1_dummies
```

```python
A_categories_2_dummies = pd.get_dummies(data.A_cat_suf_insuf, drop_first=True, prefix='A_cat2', dtype=int)
A_categories_2_dummies
```
